```python
# automate email retrieval for every name in the list
counter = 1
for index, row in tqdm(contact_list_no_email.iterrows(), total=len(contact_list_no_email), unit="rows"):
    # create base url string
    base_url = "https://api.hunter.io/v2/email-finder?company="
    # store the current company
    current_company = row['companyName']
    # add back to base url
    base_url += str(current_company) + "&full_name="
    # store current full name
    current_full_name = str(row['firstName']) + "+" + str(row['lastName'])
    # add back to base url
    base_url += current_full_name + "&api_key=" + hunter_api
    # create request to Hunter API for current row value
    current_hunter_request = requests.get(base_url)
    
    # store email in the dataframe
    try:
        contact_list_no_email['email'][index] = current_hunter_request.json()['data']['email']
    except:
        contact_list_no_email['email'][index] = ""
        
    counter += 1
    if counter > 1000:
        break
```

```python
# store the final results back to csv
contact_list_no_email.to_csv('data_files/Director_DS_LinkedIN_second_hunterio_run.csv')
```
